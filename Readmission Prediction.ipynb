Handeling Missing Values in diagnosis:

since the missing data is in secondary diagnosis I am assuming that there was no other health issue reported so I am replacing empty spaces as "None"

```python
diagnoses["secondary_diagnoses"]=diagnoses["secondary_diagnoses"].fillna("None")
```

```python
diagnoses["secondary_icd10s"].dtype
```

```python
diagnoses["secondary_icd10s"]=diagnoses["secondary_icd10s"].fillna("None")
```

```python
diagnoses.isna().sum()
```

diagnosis dataset missing values handeled

since the missing values in days_to_readmission(in outcome) so I am assuming that the patient doesn't need a readmission so I am replacing the empty spaces with -1

```python
outcomes["days_to_readmission"]=outcomes["days_to_readmission"].fillna(-1)
```
